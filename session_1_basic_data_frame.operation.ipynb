### **Step-9: Rename the column**

```python
mapper: dictionary={'City':'city'}
index:
columns:
axis:
copy:
inplace:
```

```python
import pandas as pd
names=['Ramesh','Suresh','Satish']
age=[21,22,24]
df=pd.DataFrame(zip(names,age),
                index=[1,2,3],
                columns=['Names','Age'])
df['City']=['Hyd','Pune','Blr']
df

# I want to change City column: city
```

```python
dict1={'City':'city'}
df.rename(dict1)
```

```python
dict1={'City':'city'}
df.rename(dict1,axis=1)
```

```python
dict1={'City':'city'}
df.rename(columns=dict1)
```

```python
dict1={'City':'city'}
df.rename(columns=dict1,
          inplace=True)
df
```

```python
# can you change 2 to 'B'
dict1={2:'B'}
df.rename(dict1,
         inplace=True)
```

```python
df
```

```python
dict1={2:'B'}
df.rename(index=dict1)
```

```python
dict1={2:'B'}
df.rename(dict1,axis=0)
```
